# Noise cancellation using LMS, a simple example

```python
# Contributors: Arindam Jati, Kuan-Wen Huang
%matplotlib inline
%reload_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from lms import LMS

M = 12  # Number of filter taps in adaptive filter
step = 0.1  # LMS Step size
L = 500 #optional argument to plot a part of the signal
```

### Data signal: $ \sin(2\pi 0.055)$

```python
s = np.sin(2*np.pi*0.055*np.linspace(0,999,1000))

#plot
plt.figure(figsize=(12,8))
plt.plot(s[:L]); plt.title('s: data signal');plt.xlabel("samples, n");
```

### Noise signal: $A_0 \cos(2\pi \omega_0 + \phi_0)$ 

```python
noise = 0.2 * np.cos(2*np.pi*0.5*np.linspace(0,999,1000)+0.6) #w0=0.5, A0=0.2, phi0=0.6
d = s + noise;
plt.figure(figsize=(12,8))
plt.plot(d[:L]); plt.title('Primary input / obervation signal: d = s + noise');
plt.xlabel("samples, n");
```

### Reference input / correlated noise signal: $A \cos(2\pi \omega_0 + \phi)$

```python
corr_noise = 0.5 * np.cos(2*np.pi*0.5*np.linspace(0,999,1000)+2.5) #w0=0.5, A=0.5, phi=2.5
plt.figure(figsize=(12,8))
plt.plot(corr_noise[:L]); plt.title('refernce noise')
plt.xlabel("samples, n");
plt.show()
```

### LMS test

```python
lms = LMS(L=M, step=step)
e,y,w = lms.adapt(corr_noise, d, normalize=True)
```

### plot the error signal, should converge to the data signal

```python
plt.figure(figsize=(12,8))
plt.plot(y[:L]); plt.title('y: filter output')
plt.xlabel("samples, n");
plt.show()

plt.figure(figsize=(12,8))
plt.plot(e[:L]); plt.title('e: final error signal, should converge to data signal')
plt.xlabel("samples, n");
plt.show()
```
